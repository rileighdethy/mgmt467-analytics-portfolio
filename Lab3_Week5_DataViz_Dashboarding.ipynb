### Q1 ‚Äî Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

üìù **Your response:**
*   **Exploratory Visualization:**
    *   **Audience:** Primarily the analyst or data scientist.
    *   **Purpose:** To understand the data, find patterns, identify outliers, and generate hypotheses. It's about discovery.
    *   **Level of polish:** Often rough and quick. Focus is on speed and flexibility to try different views.
    *   **Iteration:** High. You'll likely create many different visualizations as you explore the data.
*   **Explanatory Visualization:**
    *   **Audience:** A specific audience (e.g., executives, clients, the public).
    *   **Purpose:** To communicate findings, tell a story, and support a specific conclusion or recommendation. It's about communication.
    *   **Level of polish:** High. Visualizations are refined, clearly labeled, and designed for easy understanding by the target audience.
    *   **Iteration:** Lower. Once the key insights are found, the focus is on crafting the best possible visual to convey them effectively.

### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

üìù **Your response (template):**
1.  **Bar:**
    *   Question archetype: Comparison (comparing values across different categories).
    *   Data shape: One categorical variable and one quantitative variable.
    *   Pitfalls: Can be misleading if the y-axis doesn't start at zero; too many bars can be cluttered.
    *   Alternatives: Column chart (vertical bars), stacked bar chart, dot plot.
2.  **Line:**
    *   Question archetype: Trend over time, progression, or change over a continuous variable.
    *   Data shape: One continuous variable (often time) and one or more quantitative variables.
    *   Pitfalls: Can imply continuity where none exists; multiple lines can be hard to distinguish.
    *   Alternatives: Area chart, scatter plot with connected points.
3.  **Scatter:**
    *   Question archetype: Relationship or correlation between two quantitative variables.
    *   Data shape: Two quantitative variables. Can include a third quantitative variable for point size or a categorical variable for color/shape.
    *   Pitfalls: Overplotting with large datasets; can imply causality where only correlation exists.
    *   Alternatives: Bubble chart (for three quantitative variables), hexbin plot (for density with large datasets).
4.  **Histogram:**
    *   Question archetype: Distribution of a single quantitative variable.
    *   Data shape: One quantitative variable.
    *   Pitfalls: The choice of bin size can significantly alter the appearance of the distribution; does not show individual data points.
    *   Alternatives: Density plot, box plot.
5.  **Map:**
    *   Question archetype: Geographic patterns or spatial distribution.
    *   Data shape: Geographic data (latitude/longitude, regions, etc.) and one or more quantitative or categorical variables to represent on the map.
    *   Pitfalls: Can be distorted by map projections; may require specialized software or libraries; interpreting color scales can be tricky.
    *   Alternatives: Choropleth map (regions colored by value), symbol map (points sized/colored by value).

### Q3 ‚Äî Airline Executive Dashboard KPIs
**Prompt:** *"Your goal is to create a dashboard for an airline executive. Brainstorm 5 Key Performance Indicators (KPIs) that should be on this dashboard. For each KPI, suggest the best chart type to visualize it."*

Suggested KPI ideas to consider (customize them): On-Time Rate, Average Delay Minutes, Cancellation Rate, Load Factor, Yield (Revenue per Passenger Mile), Completion Factor, Net Promoter Score (if available), Revenue, Cost per Available Seat Mile (CASM).

1.  KPI: **On-Time Rate**
    *   Why it matters: Indicates the percentage of flights arriving on time, a key driver of customer satisfaction and operational efficiency.
    *   Best chart: Gauge or Scorecard (for overall rate), Bar Chart (by carrier or route), Line Chart (trend over time).
    *   Drill-down fields: Carrier, Origin, Destination, Month, Day of Week.
2.  KPI: **Average Arrival Delay Minutes**
    *   Why it matters: Measures the typical delay experienced by passengers, directly impacting customer experience and potential connection issues.
    *   Best chart: Gauge or Scorecard (for overall average), Bar Chart (by carrier or route), Line Chart (trend over time), Histogram (distribution of delays).
    *   Drill-down fields: Carrier, Origin, Destination, Month, Day of Week, Time of Day.
3.  KPI: **Load Factor**
    *   Why it matters: Represents the percentage of available seats that are filled with passengers, a crucial metric for revenue optimization and aircraft utilization.
    *   Best chart: Gauge or Scorecard (for overall average), Bar Chart (by route or aircraft type), Line Chart (trend over time).
    *   Drill-down fields: Route (Origin-Dest), Carrier, Aircraft Type, Month, Day of Week.
4.  KPI: **Revenue (or Yield)**
    *   Why it matters: Tracks the income generated by flights, a fundamental measure of financial performance. Yield (Revenue per Passenger Mile) provides a normalized view of revenue generation efficiency.
    *   Best chart: Scorecard (for total revenue), Line Chart (trend over time), Bar Chart (by route or carrier).
    *   Drill-down fields: Route (Origin-Dest), Carrier, Month, Fare Class.
5.  KPI: **Cost per Available Seat Mile (CASM)**
    *   Why it matters: Measures the cost of operating each seat over one mile, a key indicator of operational efficiency and cost control.
    *   Best chart: Scorecard (for overall average), Line Chart (trend over time), Bar Chart (by carrier or aircraft type).
    *   Drill-down fields: Carrier, Aircraft Type, Month, Cost Category (e.g., Fuel, Labor).

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

````python
# Prompt C: Delay Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=True)

plt.title('Distribution of Arrival Delays', fontsize=16)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(0, color='red', linestyle='--', label='On Time (0 min delay)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Calculate and annotate percentage of flights within [-5, 15] minutes
on_time_flights = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)].shape[0]
total_flights = df_flights.shape[0]
on_time_percentage = (on_time_flights / total_flights) * 100

plt.text(0.95, 0.95, f'Flights within [-5, 15] min: {on_time_percentage:.1f}%',
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes,
         bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
````

````python
# Prompt A: On-Time Rate by Carrier (Bar Chart)
on_time_rate = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=on_time_rate.index, y=on_time_rate.values, palette='viridis', hue=on_time_rate.index, legend=False)

plt.title('On-Time Rate by Carrier', fontsize=14)
plt.xlabel('Carrier', fontsize=12)
plt.ylabel('On-Time Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
````

````python
# Prompt B: Average Arrival Delay over Time (Line Chart)
daily_avg_delay = df_flights.groupby('date')['arr_delay_min'].mean()
rolling_avg_delay = daily_avg_delay.rolling(window=7).mean()

plt.figure(figsize=(14, 7))
plt.plot(daily_avg_delay.index, daily_avg_delay.values, label='Daily Average Delay', alpha=0.7)
plt.plot(rolling_avg_delay.index, rolling_avg_delay.values, label='7-Day Rolling Average', color='red')

plt.title('Average Arrival Delay Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Arrival Delay (minutes)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Prompt C: Delay Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=True)

plt.title('Distribution of Arrival Delays', fontsize=16)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(0, color='red', linestyle='--', label='On Time (0 min delay)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Calculate and annotate percentage of flights within [-5, 15] minutes
on_time_flights = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)].shape[0]
total_flights = df_flights.shape[0]
on_time_percentage = (on_time_flights / total_flights) * 100

plt.text(0.95, 0.95, f'Flights within [-5, 15] min: {on_time_percentage:.1f}%',
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes,
         bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
````

````python
# Prompt D: Load Factor by Route (Top 15) (Horizontal Bar)
route_load_factor = df_flights.groupby(['origin', 'dest'])['load_factor'].mean().sort_values(ascending=False).head(15)

# Combine origin and destination into a single string for plotting
route_labels = [f'{origin} - {dest}' for origin, dest in route_load_factor.index]

plt.figure(figsize=(12, 8))
sns.barplot(x=route_load_factor.values, y=route_labels, palette='viridis', orient='h', hue=route_labels, legend=False)

plt.title('Top 15 Routes by Average Load Factor', fontsize=16)
plt.xlabel('Average Load Factor', fontsize=12)
plt.ylabel('Route (Origin - Dest)', fontsize=12)
plt.xlim(0, 1)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
# Prompt E: Cost vs. Revenue with CASM Coloring (Scatter)
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df_flights, x='cost_usd', y='revenue_usd', hue='casm', palette='viridis', alpha=0.6)

plt.title('Cost vs. Revenue (Colored by CASM)', fontsize=16)
plt.xlabel('Cost (USD)', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Add y=x reference line
max_val = max(df_flights['cost_usd'].max(), df_flights['revenue_usd'].max())
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='Cost = Revenue')

plt.legend(title='CASM')
plt.tight_layout()
plt.show()
````
